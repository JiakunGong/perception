```python
bib = []
with open('a.bib', 'r') as f:
  bib = f.readlines()
  
outbib = []
```

```python
fixcounter = 0
for l in bib:
  
  output = l
  
  if l.find('Author') != -1 or l.find('author') != -1:

    authors = l.split('},')[0]
#     print authors
    
#     authors = l.split('{')[-1].split('}')[0]
    if authors.find(',') == -1: #and authors.find('.') == -1:
#       continue
#       print authors
  
      # we now have stuff like this:
      # Author = {J Lichtman and J Livet and J Sanes
      # and like this
      # Author = {Yuri Boykov and Olga Veksler and Ramin Zabih
      
      authors = authors.split('{')[-1]
      

      if authors.find('}') != -1 or authors.startswith('NSF'):
        print 'SKIPPING', authors
        continue

#       print authors
      isolated_names = authors.split('and')
      new_names = []
      ctr = 0
      for j,i in enumerate(isolated_names):
#         print i
        split_name = i.strip().split(' ')
        
        if len(split_name[0].strip()) == 1:
        
          ctr += 1
          # this is N Kasthuri and J Sanes.. but should be Kasthuri, N. and Sanes, J.
          new_name = split_name[1].strip() + ', ' + split_name[0].strip() + '.'
          new_names.append(new_name)
          
#         print split_name, len(split_name)
        if len(split_name) == 2:
          if len(split_name[1].strip()) == 1 or len(split_name[1].strip()) == 2:

            ctr += 1

            # this is N Kasthuri and J Sanes.. but should be Kasthuri, N. and Sanes, J.
            new_name = split_name[0].strip() + ', ' + split_name[1].strip() + '.'
            new_names.append(new_name)
            
          else:
            ctr += 1
            # regular case like Johanna Beyer and Daniel Haehn
            new_name = split_name[1].strip() + ', ' + split_name[0].strip()[0] + '.'
            new_names.append(new_name)
            
        if len(split_name) == 3:
          # Jeff W. Lichtman
          ctr += 1
          # regular case like Johanna Beyer and Daniel Haehn
          new_name = split_name[2].strip() + ', ' + split_name[0].strip()[0] + '.' + split_name[1].strip()[0] + '.'
          new_names.append(new_name)
        
#       print ctr, j
      if ctr != j+1:
        new_names = [] # ignore J. Lu etc.. short last names by looking at all authors
          
          
      if len(new_names) != 0:
        fixcounter += 1
        # we fixed something
        print authors
        print 'is now'
        print ' and '.join(new_names)
        print ''
        output = 'Author = {' + ' and '.join(new_names) + '},'
        
  outbib.append(output)
```
