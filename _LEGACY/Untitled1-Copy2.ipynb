```python
bib = []
with open('a.bib', 'r') as f:
  bib = f.readlines()
  
```

```python
outbib = []
fixcounter = 0
for l in bib:
  
  output = l
  
  
  if l.find('Author') != -1 or l.find('author') != -1:

    authors = l[l.find("{")+1:l.rfind("}")]
    
    # we only care if ' and ' is in the authors
    if authors.lower().find(' and ') == -1:
      print 'SKIPPING', authors
      continue
      
    splitted_authors = authors.split(' and ')
    if len(splitted_authors) == 1:
      splitted_authors = authors.split(' AND ')
    
    new_names = []
    
    for s in splitted_authors:
    
      # case 1: there is no comma (Daniel Haehn, Daniel F. Haehn, Daniel F. G. H. Haehn)
      if s.find(',') == -1:
        # lastname is always last
        lastname = s.split(' ')[-1]
        firstnames = s.split(' ')[:-1]
        
        converted_firstnames = ''
        if len(firstnames) > 0:
          for f in firstnames:
            converted_firstnames += f[0] + '.'
        new_name = lastname + ', ' + converted_firstnames.strip()
        print s, firstnames, new_name
        new_names.append(new_name)
        
      # case 2: there is a comma, just abbreviate the names
      if s.find(',') != -1:
        # lastname is always first
        lastname = s.split(',')[0]
        firstnames = s.split(',')[1:]
        converted_firstnames = ''
#         print firstnames
        if len(firstnames) > 0 and firstnames[0] != '':
#           print firstnames
          for f in firstnames:
            converted_firstnames += f.strip()[0] + '.'
        new_name = lastname + ', ' + converted_firstnames.strip()
        
#         print s, '-->', new_name

        new_names.append(new_name)
    
    if new_names:
#       print splitted_authors
#       print new_names
#       print
      output = '  author = {' + ' and '.join(new_names) + '},\n'
      fixcounter += 1

  outbib.append(output)
    
```

```python
with open('new.bib', 'w') as f:
  f.write(''.join(outbib))
```
